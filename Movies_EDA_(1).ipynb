## Import Libraries

```python
import numpy as np
import pandas as pd
```

## Reading Data

```python
df = pd.read_csv('/content/imdb_movies.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

## Drop Unnecessary Columns

```python
data = df.drop(['certificate', 'runtime'], axis=1)
```

```python
data.head()
```

## Rename Columns in a suitable form

```python
data.columns = data.columns.str.strip().str.lower().str.replace(' ', '_')
data.head()
```

```python
data.columns
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace=True)
```

```python
data.isnull().sum()
```

```python
data.describe()
```

## What is the minimum and maximum ratings ?


```python
data['rating'].min()
```

```python
data['rating'].max()
```

## What are the movies with rating > 9 ?

```python
data[data['rating'] > 9]
```

```python
data.head()
```

## Top 10 movies per metascore and user rating

```python
data.sort_values(by=['metascore', 'rating'], ascending=False).head(10)
```

## Top 10 genres

```python
data['genre'].value_counts().head(10)
```

```python
genre_ratings = data.groupby('genre')['rating'].mean().sort_values(ascending=False).head(10)
genre_ratings
```

## Top 10 Directories

```python
director_ratings = data.groupby('director')['rating'].mean().sort_values(ascending=False).head(10)
director_ratings
```

```python
data['director'].value_counts().head(10)
```

## Top 10 First Actors

```python
data['actor_1'].value_counts().head(10)
```

```python
actor_ratings = data.groupby('actor_1')['rating'].mean().sort_values(ascending=False).head(10)
actor_ratings
```

## Top 10 Second Actors

```python
actor_ratings = data.groupby('actor_2')['rating'].mean().sort_values(ascending=False).head(10)
actor_ratings
```
